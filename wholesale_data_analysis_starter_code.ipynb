## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df_with_cost=(df
              .pipe(lambda x: x.assign(all_item_price=x["qty"]*x["unit_price"])))

#look at the data
df_with_cost.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.

#create shipping_unit_price function
def weight_unit_price(weight):
    if(weight<=50):
        weight_unit_price=10.00
    else:
        weight_unit_price=7.00
    return weight_unit_price

#continuing with the pipe syntax
df_with_cost=(df
              #from previous step
              .pipe(lambda x: x.assign(all_item_price=x["qty"]*x["unit_price"])) 
              # new to this step
              .pipe(lambda x: x.assign(total_weight=x["qty"]*x["unit_weight"]))
              .pipe(lambda x: x.assign(shipping_unit_price=x["total_weight"].apply(weight_unit_price)))
              .pipe(lambda x: x.assign(shipping_price=x["shipping_unit_price"]*x["total_weight"]))
)

#look at the data
df_with_cost.head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%

#define function for total cost to make total cost column cleaner
def total_cost(base_cost, shipping_cost,sales_tax):
    cost_no_tax=base_cost+shipping_cost
    cost_of_tax=cost_no_tax*sales_tax
    total_cost=cost_no_tax+cost_of_tax
    return total_cost

#continuing with the pipe syntax
df_with_cost=(df
              #from previous steps
              .pipe(lambda x: x.assign(all_item_price=x["qty"]*x["unit_price"])) 
              .pipe(lambda x: x.assign(total_weight=x["qty"]*x["unit_weight"]))
              .pipe(lambda x: x.assign(shipping_unit_price=x["total_weight"].apply(weight_unit_price)))
              .pipe(lambda x: x.assign(shipping_price=x["shipping_unit_price"]*x["total_weight"]))
              #new to this step
              .pipe(lambda x: x.assign(sales_tax=.0925))
              #needed bing copilot to send multiple arguments but did not want to create intermediate columns
              .pipe(lambda x: x.assign(total_price=x.apply(lambda row: 
              total_cost(row["all_item_price"],row["shipping_price"],row["sales_tax"]),axis=1)))
              )


#look at the data
df_with_cost.head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

#continuing with the pipe syntax
df_with_cost=(df
              #from previous steps
              .pipe(lambda x: x.assign(all_item_price=x["qty"]*x["unit_price"])) 
              .pipe(lambda x: x.assign(total_weight=x["qty"]*x["unit_weight"]))
              .pipe(lambda x: x.assign(shipping_unit_price=x["total_weight"].apply(weight_unit_price)))
              .pipe(lambda x: x.assign(shipping_price=x["shipping_unit_price"]*x["total_weight"]))
              .pipe(lambda x: x.assign(sales_tax=.0925))
              #needed bing copilot to send multiple arguments but did not want to create intermediate columns
              .pipe(lambda x: x.assign(total_price=x.apply(lambda row: 
              total_cost(row["all_item_price"],row["shipping_price"],row["sales_tax"]),axis=1)))
            #new to this step
              .pipe(lambda x: x.assign(all_item_cost=x["qty"]*x["unit_cost"]))
              .pipe(lambda x: x.assign(line_cost=x.apply(lambda row: 
              total_cost(row["all_item_cost"],row["shipping_price"],0),axis=1)))
              )

#look at the data
df_with_cost.head()


```

```python
# Create a column for the profit of each line using line cost and line price
#continuing with the pipe syntax

#define a function to get 2 decimal places
def format_cost(value):
        return round(value,2)

df_with_cost=(df
              #from previous steps
              .pipe(lambda x: x.assign(all_item_price=x["qty"]*x["unit_price"])) 
              .pipe(lambda x: x.assign(total_weight=x["qty"]*x["unit_weight"]))
              .pipe(lambda x: x.assign(shipping_unit_price=x["total_weight"].apply(weight_unit_price)))
              .pipe(lambda x: x.assign(shipping_price=x["shipping_unit_price"]*x["total_weight"]))
              .pipe(lambda x: x.assign(sales_tax=.0925))
              #needed bing copilot to send multiple arguments but did not want to create intermediate columns
              .pipe(lambda x: x.assign(total_price=x.apply(lambda row: 
              total_cost(row["all_item_price"],row["shipping_price"],row["sales_tax"]),axis=1)))
              .pipe(lambda x: x.assign(all_item_cost=x["qty"]*x["unit_cost"]))
              .pipe(lambda x: x.assign(line_cost=x.apply(lambda row: 
              total_cost(row["all_item_cost"],row["shipping_price"],0),axis=1)))
              #new to this step
              .pipe(lambda x: x.assign(profit=x["total_price"]-x["line_cost"]))
              )

#formatting columns
columns_to_format=["all_item_price","shipping_unit_price","shipping_price","total_price","all_item_cost","line_cost","profit"]
df_with_cost_formatted=df_with_cost
df_with_cost_formatted[columns_to_format]=df_with_cost[columns_to_format].map(format_cost)

#look at the data
df_with_cost_formatted.head()



```

Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

#limit clients to top 5 list and then sum the total price for each
top5_spend=(df_with_cost_formatted
            .pipe(lambda df: df[df["client_id"].isin(top5_clients_list)])
            .pipe(lambda df: df.groupby("client_id")["total_price"].sum().      reset_index())
            #needed some chatgpt help with getting the sort in same order as the client list
            .pipe(lambda df: df.sort_values(by="client_id", key=lambda x: pd.Categorical(x, categories=top5_clients_list,ordered=True)))           
            )

top5_spend

#the line below does work but still not in the order of top5_clients_list.
#df_with_cost_formatted[df_with_cost_formatted["client_id"].isin(top5_clients_list)].groupby("client_id")["total_price"].sum()

```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit.
 
top5_spend=(df_with_cost_formatted
            .pipe(lambda df: df[["client_id","qty","shipping_price","total_price","profit"]])
            .pipe(lambda df: df[df["client_id"].isin(top5_clients_list)])
            .pipe(lambda df: df.groupby("client_id").sum().reset_index())
            #needed some chatgpt help with getting the sort in same order as the client list
            .pipe(lambda df: df.sort_values(by="client_id", key=lambda x: pd.Categorical(x, categories=top5_clients_list,ordered=True)))           
            )

top5_spend
```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns. 
money_cols=["shipping_price","total_price","profit"]
money_cols

# Define a function that converts a dollar amount to millions.
def currency_format_millions(col):
    col_millions=col/1000000
    return col_millions

# Apply the currency_format_millions function to only the money columns.
top5_spend_pres=top5_spend.copy()
top5_spend_pres[money_cols]=top5_spend_pres[money_cols].apply(currency_format_millions)

# Rename the columns to reflect the change in the money format. 
top5_spend_pres.rename(columns={"client_id":"Client ID", "qty":"Units","shipping_price":"Shipping (millions)","total_price":"Total Revenue (millions)","profit":"Total Profit (millions)"})

#make multi row columns names
top5_spend_pres.columns=pd.MultiIndex.from_tuples([
    ("Client","ID"),
    ("Units",""),
    ("Shipping", "(millions)"),
    ("Total Revenue", "(millions)"),
    ("Total Profit", "(millions)")
])

top5_spend_pres
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.

#top5_spend_pres_sort=top5_spend_pres.sort_values
top5_spend_pres_sort=top5_spend_pres.sort_values(('Total Profit','(millions)'),ascending=False)

top5_spend_pres_sort

```
